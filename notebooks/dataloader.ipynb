```python
import sys
sys.path.append("..")

import torch
from torch import nn
import matplotlib.pyplot as plt
import torchvision.utils

from lmc.utils.setup_training import setup_loader
from lmc.config import DataConfig

%load_ext autoreload
%autoreload 2
```

```python
conf = DataConfig(dataset="cifar10", hflip=True, path="../data", batch_size=128, test_batch_size=8)
seed = 42
```

```python
train_loader = setup_loader(conf, train=True, eval=False, loader_seed=seed)
train_loader2 = setup_loader(conf, train=True, eval=False, loader_seed=seed)
test_loader = setup_loader(conf, train=False, eval=True, loader_seed=seed)
```

```python
def visualize_batch(dataloader, n_images: int = 8):
    # Get a batch of images and labels from the dataloader
    images, labels = next(iter(dataloader))
    images = images[:n_images]
    labels = labels[:n_images]
    
    # Create a grid of images
    grid_img = torchvision.utils.make_grid(images, nrow=8)
    
    # Display the images
    plt.figure(figsize=(10,10))
    plt.imshow(grid_img.permute(1, 2, 0))  # Permute to move the color channel to the last position
    plt.title(f"Labels: {labels.tolist()}")
    plt.axis("off")
    plt.show()

print("Loader 1, hflip")
visualize_batch(train_loader)
visualize_batch(train_loader)
print("Loader 1, hflip")
visualize_batch(train_loader2)
visualize_batch(train_loader2)
visualize_batch(test_loader)
```
